```python
# 1. Reload the file fresh
df = pd.read_excel(r'F:\Money Forcasting\Dataset\Components of Money Stock.xlsx', header=3)

# 2. If 'Unnamed: 0' exists, drop it
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# 3. Rename columns using the list you already created
# Make sure the length of this list exactly matches the number of columns in your df
df.columns = [
    'Year', 'Month', 'Currency_in_Circulation', 'Cash_with_Banks',
    'Currency_with_Public', 'Other_Deposits_RBI', 'Bankers_Deposits_RBI',
    'Demand_Deposits', 'Time_Deposits', 'Time_Deposits_Excluding_Merger',
    'Reserve_Money_M0', 'Narrow_Money_M1', 'Broad_Money_M3',
    'Broad_Money_M3_Excluding_Merger'
]

# 4. Now perform the conversion
cols_to_convert = ['Currency_in_Circulation', 'Cash_with_Banks', 
                   'Currency_with_Public', 'Other_Deposits_RBI', 
                   'Bankers_Deposits_RBI', 'Demand_Deposits', 
                   'Time_Deposits', 'Time_Deposits_Excluding_Merger', 
                   'Reserve_Money_M0', 'Narrow_Money_M1', 
                   'Broad_Money_M3', 'Broad_Money_M3_Excluding_Merger']

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print(df.info())
```
